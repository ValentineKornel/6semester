```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#1
array = np.random.randint(1, 11, size=(4, 5))
print ("Sourse array:\n", array)

array1, array2 = np.array_split(array, 2)
print("Array 1:\n", array1)
print("Array 2:\n", array2)

target_value = 6
matches = np.where(array1 == target_value)
print(f"Indexes of found values {target_value} in first array: {matches}")

count = np.count_nonzero(array1 == target_value);
print(f"Amount of elements equals to {target_value}: {count}")

#2
data = np.array([10, 20, 30, 40, 50])
series = pd.Series(data);
print("Series:\n", series)


print("Series + 5:\n", series + 5)
print("Series * 2:\n", series * 2)
print("Average:\n", series.mean())
print("Max value:\n", series.max())
print("Min value:\n", series.min())

array = np.random.randint(1, 101, size=(5, 3))
df = pd.DataFrame(array, columns=['A', 'B', 'C'])
df.at[2, 'C'] = 50
print("\nDataFrame:\n", df)

df = df.drop(index=1);
print("\nDataFrame after deleting the string:\n", df)

df = df.drop(columns=['B'])
print("\nDataFrame after deletin the column:\n", df)

print("\nDataFrame shape: ", df.shape)


target_value = 50
matches = df[df == target_value].stack().index.tolist()
print(f"\nIndexes of element equals to {target_value}: ", matches)


#3

url = "https://raw.githubusercontent.com/ValentineKornel/config/refs/heads/main/StudentPerformanceFactors.csv"
df = pd.read_csv(url)
print(df.head())

df['Exam_Score'].max()
max_index = df['Exam_Score'].idxmax()
print("Индекс максимального элемента:", max_index)
df.at[max_index, 'Exam_Score'] = 100


plt.hist(df['Exam_Score'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel("Exam Score")
plt.ylabel("Frequency")
plt.title("Histogram of Exam Scores")
plt.show()

median_score = df['Exam_Score'].median()
mean_score = df['Exam_Score'].mean()

print(f"Медиана: {median_score}")
print(f"Среднее значение: {mean_score}")


plt.boxplot(df['Exam_Score'])
plt.ylabel("Exam Score")
plt.title("Box Plot of Exam Scores")
plt.show()

df['Exam_Score'].describe()

grouped = df.groupby("Parental_Involvement")["Exam_Score"].mean()
print(grouped)
grouped = df.groupby("Sleep_Hours")["Exam_Score"].mean()
print(grouped)
grouped = df.groupby("Hours_Studied")["Exam_Score"].mean()
print(grouped)
grouped = df.groupby("Attendance")["Exam_Score"].mean()
print(grouped)
grouped = df.groupby("Access_to_Resources")["Exam_Score"].mean()
print(grouped)
grouped = df.groupby("Motivation_Level")["Exam_Score"].mean()
print(grouped)
grouped = df.groupby("Peer_Influence")["Exam_Score"].mean()
print(grouped)

```
